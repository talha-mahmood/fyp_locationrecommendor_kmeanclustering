```python

#imports
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')



import statsmodels.api as sm
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
from sklearn.cluster import KMeans

df = pd.read_csv('./locations.csv', sep=',', encoding='ISO-8859-1')

#geopy in action
country = 'Pakistan'
city_names = df['name']

longitude = []
latitude = []

l2 = df[['longitude', 'latitude']]

# Implementing the Elbow Method
wcss = []  # within-cluster sum of squares

for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', random_state=42)
    kmeans.fit(l2)
    wcss.append(kmeans.inertia_)

# Plotting the Elbow Method graph
plt.plot(range(1, 11), wcss)
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()

# Choose the optimal number of clusters based on the Elbow Method
optimal_clusters =5

kmeans = KMeans(n_clusters=optimal_clusters, init='k-means++', random_state=42)
identified_clusters = kmeans.fit_predict(l2)
df['loc_clusters'] = identified_clusters
df.to_csv('pakistanlocations.csv')

#finishing touches!
input_city = input("Enter a city name: ")

# Check if the city exists in the dataset
if input_city in df['name'].values:
    cluster = df.loc[df['name'] == input_city, 'loc_clusters'].iloc[0]
    cities = df.loc[df['loc_clusters'] == cluster, 'name']
    for city in cities:
        if city != input_city:
            print(city)
else:
    print("City not found in the dataset.")
```
